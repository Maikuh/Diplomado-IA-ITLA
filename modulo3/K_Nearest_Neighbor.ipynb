## Importar librerias


```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

## TODO: Importar estimador del KNN
from sklearn.neighbors import NearestNeighbors
```

## Importar datos

```
# TODO: Importar nombre de la libreria de Scikit learn
from sklearn.datasets import load_iris
data = load_iris()

## TODO: Importar datos a respectivas variables
X = data.data
y = data.target
```

### Tipo de datos en Python de X & y
Observar que tipo de datos son X & y provenientes de Scikit-learn.

Usar función interna de Python *type* para ver el tipo de variable que son.

```
## TODO: Imprimir el TIPO de variable que es X & y
print("Tipo de X: ", type(X))
print("Tipo de y: ", type(y))
```

### Explorando con Pandas
Uso de la librería de Pandas para explorar los datos.

Con el método *DataFrame* estamos creando un *DataFrame* en Pandas. *DataFrame* es una tabla 2D. Compuesta por filas y columnas.

Primer parametros son las filas, y el segundo es el nombre de las columnas.

Método **head()** imprime los primeros 5 ejemplos/observaciones.

```
df = pd.DataFrame(X, columns = data.feature_names)
df.head()
```

```
## TODO: Usa método para DESCRIBIR un Dataframe de Pandas
df.describe()
```

#### Graficando desde Pandas
A continuación se grafica con Pandas usando el método *plotting* y *scatter_matrix*. Existen multiples tipos de gráficos incluidos ya en Pandas.

```
_ = pd.plotting.scatter_matrix(df, c = y, figsize = [10,10], s=50, marker='C')
```

## Separación en base de Entrenamiento y Prueba

```
## TODO: Importar separador entre entrenamiento y prueba
from sklearn.model_selection import train_test_split

## TODO: Crear los diferentes sets de datos
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 20)
```

## Instanciando el Modelo
Instanciar tres modelos con tres diferentes valores de vecinos.

Parametro a cambiar:

**n_neighbors**

Instanciarlos con tres nombres diferentes. Por ejemplo:

"**knn3**" si usa tres vecinos.

```
from sklearn.neighbors import KNeighborsClassifier

## TODO: Instanciar 3 modelos con diferentes vecinos. Un vecino debe ser igual a 3, y los otros libres.
knn3 = KNeighborsClassifier(n_neighbors=3)
knn10 = KNeighborsClassifier(n_neighbors=10)
knn50 = KNeighborsClassifier(n_neighbors=50)
```

## Fitting los Tres Modelo

```
## TODO: Fit/Entrena los 3 modelos
knn3.fit(X_train, y_train)
knn10.fit(X_train, y_train)
knn50.fit(X_train, y_train)
```

## Prediciendo 

```
## TODO: Predecir con los 3 modelos...
y_pred3 = knn3.predict(X_test)
y_pred10 = knn10.predict(X_test)
y_pred50 = knn50.predict(X_test)
```

## Evaluando los modelos

```
## TODO: 
print('Modelo KNN3:', knn3.score(X_test, y_test))
print('Modelo KNN10:', knn10.score(X_test, y_test))
print('Modelo KNN50:', knn50.score(X_test, y_test))
```
