# Sin Supervisión: Clustering

## Cargar los datos


```python
# Importando libreria para crear datos aleatorios
from sklearn.datasets.samples_generator import make_blobs

# Definiendo centros
centers = [[1,1],[1.5,1.5],[2,2]]

# Creando nuestro X & y
X,y = make_blobs(n_samples=100, centers = centers, cluster_std=0.1)

# Gráficando nuestros datos
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1])
```

## Cargar modelo

```python
# Importando modelo
from sklearn.cluster import KMeans

# Instanciar modelo con parametros
cluster = KMeans(n_clusters=3, random_state=10)
```

## Entrenando Modelo

```python
cluster.fit(X)
```

## Evaluando Modelo

```python
# Imprimiendo al cluster que fue asignado cada punto
cluster.labels_
```

```python
# Imprimiendo los centroids de los clusters
cluster.cluster_centers_
```

```python
# Gráficando lo actual contra lo predicho
plt.subplot(1,2,1)
plt.scatter(X[:,0],X[:,1],c=y)
plt.title("Actual")
plt.subplot(1,2,2)
plt.scatter(X[:,0],X[:,1],c=cluster.labels_)
plt.title("Predición")
```

## Ejemplo con datos de Iris

```python
# Importar base de datos
from sklearn import datasets
# TODO: Importar modelo
from sklearn.cluster import KMeans

# Cargar datos
iris = datasets.load_iris()

# Separar en X & y
X, y = iris.data, iris.target

# TODO: Instanciar modelo
cluster = KMeans(n_clusters=3, random_state=10)

# TODO: Entrenar modelo
cluster.fit(X)
```

```python
# Gráficar 
a = 1 # Petal length
b = 2 # Sepal width
plt.subplot(1,2,1)
plt.scatter(X[:,a],X[:,b],c=y) 
plt.xlabel(iris.feature_names[a])
plt.ylabel(iris.feature_names[b])
plt.title("Actual")
plt.subplot(1,2,2)
plt.scatter(X[:,a],X[:,b],c=cluster.labels_)
plt.title("Predición")
```
